# **Question 1 : Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.**

SQL (Structured Query Language) commands are broadly categorized into three main types:

1.  **DDL (Data Definition Language)**
    *   **Purpose**: DDL commands are used to define, modify, or delete the structure of database objects (like tables, indexes, users). They deal with the schema of the database.
    *   **Effect**: Changes are permanent and affect the database schema directly.
    *   **Examples**:
        *   `CREATE`: Used to create new database objects.
            ```sql
            CREATE TABLE Employees (
                EmployeeID INT PRIMARY KEY,
                FirstName VARCHAR(50),
                LastName VARCHAR(50),
                DepartmentID INT
            );
            ```
        *   `ALTER`: Used to modify the structure of existing database objects.
            ```sql
            ALTER TABLE Employees
            ADD Email VARCHAR(100);
            ```
        *   `DROP`: Used to delete existing database objects.
            ```sql
            DROP TABLE Employees;
            ```
        *   `TRUNCATE`: Used to remove all records from a table, but the table structure remains.
            ```sql
            TRUNCATE TABLE Employees;
            ```

2.  **DML (Data Manipulation Language)**
    *   **Purpose**: DML commands are used for managing data within schema objects. They deal with the data stored in the database.
    *   **Effect**: Changes are not permanent until a `COMMIT` statement is executed (in transactional databases) and can be rolled back.
    *   **Examples**:
        *   `INSERT`: Used to add new rows of data into a table.
            ```sql
            INSERT INTO Employees (EmployeeID, FirstName, LastName, DepartmentID, Email)
            VALUES (1, 'John', 'Doe', 101, 'john.doe@example.com');
            ```
        *   `UPDATE`: Used to modify existing data within a table.
            ```sql
            UPDATE Employees
            SET Email = 'john.d@example.com'
            WHERE EmployeeID = 1;
            ```
        *   `DELETE`: Used to remove one or more rows from a table.
            ```sql
            DELETE FROM Employees
            WHERE EmployeeID = 1;
            ```

3.  **DQL (Data Query Language)**
    *   **Purpose**: DQL commands are used to retrieve data from the database. It is primarily used for querying data.
    *   **Effect**: These commands do not modify the database; they only retrieve data.
    *   **Examples**:
        *   `SELECT`: Used to retrieve data from one or more tables.
            ```sql
            SELECT EmployeeID, FirstName, LastName
            FROM Employees
            WHERE DepartmentID = 101;
            ```


# **Question 2 : What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.**

**Purpose of SQL Constraints:**

SQL constraints are rules enforced on data columns in a table. Their primary purpose is to limit the type of data that can go into a table, ensuring the accuracy, reliability, and integrity of the data. They are used to:

*   **Enforce Data Integrity**: Prevent the entry of invalid data into the database.
*   **Maintain Relationships**: Define and enforce relationships between tables.
*   **Improve Data Quality**: Ensure that data meets specific business rules.

**Three Common Types of Constraints:**

1.  **PRIMARY KEY Constraint**

    *   **Description**: A PRIMARY KEY is a field or combination of fields that uniquely identifies each record in a table. It cannot contain NULL values, and each value must be unique. A table can only have one primary key.
    *   **Scenario**: In an `Employees` table, `EmployeeID` would be an ideal primary key. Each employee has a unique ID, and this ID is essential for identifying individual employee records. It ensures that no two employees have the same ID and that every employee record has an ID.

    ```sql
    CREATE TABLE Employees (
        EmployeeID INT PRIMARY KEY, -- Primary Key constraint
        FirstName VARCHAR(50),
        LastName VARCHAR(50)
    );
    ```

2.  **FOREIGN KEY Constraint**

    *   **Description**: A FOREIGN KEY is a field (or collection of fields) in one table that refers to the PRIMARY KEY in another table. It establishes and enforces a link between two tables, ensuring referential integrity. This means that data in the foreign key column must exist in the primary key column of the referenced table, or be NULL.
    *   **Scenario**: Consider a `Departments` table with `DepartmentID` as its primary key, and an `Employees` table. To link an employee to their department, we would add a `DepartmentID` column to the `Employees` table and make it a foreign key referencing the `DepartmentID` in the `Departments` table. This ensures that an employee cannot be assigned to a `DepartmentID` that does not exist in the `Departments` table.

    ```sql
    CREATE TABLE Departments (
        DepartmentID INT PRIMARY KEY,
        DepartmentName VARCHAR(100)
    );

    CREATE TABLE Employees (
        EmployeeID INT PRIMARY KEY,
        FirstName VARCHAR(50),
        LastName VARCHAR(50),
        DepartmentID INT,
        FOREIGN KEY (DepartmentID) REFERENCES Departments(DepartmentID) -- Foreign Key constraint
    );
    ```

3.  **NOT NULL Constraint**

    *   **Description**: The NOT NULL constraint ensures that a column cannot have a NULL value. This means that a value must be provided for that column whenever a new record is inserted or an existing record is updated.
    *   **Scenario**: In a `Customers` table, columns like `FirstName` and `LastName` would typically have a NOT NULL constraint. It's crucial for customer records to have a name, so these fields should never be left empty (NULL).

    ```sql
    CREATE TABLE Customers (
        CustomerID INT PRIMARY KEY,
        FirstName VARCHAR(50) NOT NULL, -- NOT NULL constraint
        LastName VARCHAR(50) NOT NULL,  -- NOT NULL constraint
        Email VARCHAR(100)
    );
    ```

# **Question 3 : Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?**

In SQL, `LIMIT` and `OFFSET` are clauses used with the `SELECT` statement to control the number of rows returned by a query and to skip a certain number of rows before beginning to return the result set.

### **1. LIMIT Clause**

*   **Purpose**: The `LIMIT` clause is used to restrict the number of rows returned by the query. It specifies the maximum number of rows that the query should retrieve.
*   **Syntax**: `LIMIT count`
*   **Example**: To get only the first 5 employees:
    ```sql
    SELECT * FROM Employees
    LIMIT 5;
    ```

### **2. OFFSET Clause**

*   **Purpose**: The `OFFSET` clause is used to skip a specified number of rows from the beginning of the result set before returning the remaining rows. It is typically used in conjunction with `LIMIT` for pagination.
*   **Syntax**: `OFFSET start`
*   **Example**: To skip the first 10 employees and then get all subsequent employees:
    ```sql
    SELECT * FROM Employees
    OFFSET 10;
    ```

### **Using LIMIT and OFFSET together for Pagination**

When combined, `LIMIT` and `OFFSET` allow you to implement pagination, which means retrieving data in chunks (pages). `OFFSET` determines where to start fetching records, and `LIMIT` determines how many records to fetch from that starting point.

#### **Scenario: Retrieve the third page of results, assuming each page has 10 records.**

To get the third page, you need to:

1.  **Skip the records from the previous pages**: If each page has 10 records, then the first two pages contain `2 * 10 = 20` records. So, you need to `OFFSET` by 20.
2.  **Retrieve 10 records for the current page**: You then `LIMIT` the result to 10 records.

**Calculation for OFFSET**: `(page_number - 1) * records_per_page`
For the third page (page_number = 3) with 10 records per page:
`OFFSET = (3 - 1) * 10 = 2 * 10 = 20`

**SQL Query**: To retrieve the third page of results (records 21-30):

```sql
SELECT * FROM YourTableName
ORDER BY SomeColumn -- Important for consistent pagination
LIMIT 10 OFFSET 20;
```

**Explanation of the query:**

*   `SELECT * FROM YourTableName`: Selects all columns from the specified table.
*   `ORDER BY SomeColumn`: It is crucial to use an `ORDER BY` clause to ensure that the results are consistently ordered. Without it, the order of rows might be arbitrary, leading to unpredictable pagination results.
*   `LIMIT 10`: Retrieves a maximum of 10 rows.
*   `OFFSET 20`: Skips the first 20 rows before starting to retrieve the 10 rows specified by `LIMIT`.

# **Question 4 : What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.**

### **What is a Common Table Expression (CTE)?**

A Common Table Expression (CTE) is a temporary, named result set that you can reference within a single SQL statement (such as `SELECT`, `INSERT`, `UPDATE`, or `DELETE`). It's defined using the `WITH` clause, providing a way to break down complex queries into simpler, more readable, and manageable logical steps. A CTE exists only for the duration of the query in which it is defined.

### **Main Benefits of CTEs:**

1.  **Readability and Maintainability**: CTEs significantly improve the readability of complex queries by breaking them into logical, self-contained blocks. This makes the SQL code easier to understand, debug, and maintain.

2.  **Modularity**: They allow you to define intermediate result sets that can be reused multiple times within the same query without having to rewrite the same logic or create subqueries repeatedly.

3.  **Simplifying Complex Joins and Subqueries**: CTEs can replace complex subqueries and derived tables, making the query structure flatter and more intuitive. This is especially useful for hierarchical queries or recursive queries.

4.  **Enabling Recursion**: CTEs are essential for performing recursive queries, which are queries that refer to themselves. This is commonly used for querying hierarchical data like organizational charts or bill of materials.

5.  **Improved Performance (in some cases)**: While not always a direct performance booster, by making queries clearer and allowing the optimizer to better understand the query's intent, CTEs can sometimes lead to more efficient execution plans.

### **Simple SQL Example Demonstrating CTE Usage:**

Let's imagine we have a table `Sales` with `ProductID`, `SaleDate`, and `Amount`. We want to find the total sales for each product and then only select products that had total sales greater than 1000.

```sql
-- Sample Data (for illustration, not part of the CTE definition itself)
CREATE TABLE Sales (
    ProductID INT,
    SaleDate DATE,
    Amount DECIMAL(10, 2)
);

INSERT INTO Sales (ProductID, SaleDate, Amount) VALUES
(1, '2023-01-01', 150.00),
(1, '2023-01-02', 200.00),
(2, '2023-01-01', 500.00),
(2, '2023-01-03', 700.00),
(3, '2023-01-05', 300.00),
(1, '2023-01-03', 800.00);

-- Using a CTE to calculate total sales per product and then filter
WITH ProductTotalSales AS (
    -- This is the CTE definition
    SELECT
        ProductID,
        SUM(Amount) AS TotalSalesAmount
    FROM
        Sales
    GROUP BY
        ProductID
)
-- This is the main query that uses the CTE
SELECT
    ProductID,
    TotalSalesAmount
FROM
    ProductTotalSales
WHERE
    TotalSalesAmount > 1000;
```

**Explanation:**

1.  **`WITH ProductTotalSales AS (...)`**: This defines a CTE named `ProductTotalSales`.
2.  **`SELECT ProductID, SUM(Amount) AS TotalSalesAmount FROM Sales GROUP BY ProductID`**: Inside the CTE, we calculate the `TotalSalesAmount` for each `ProductID` from the `Sales` table.
3.  **`SELECT ProductID, TotalSalesAmount FROM ProductTotalSales WHERE TotalSalesAmount > 1000;`**: The main query then treats `ProductTotalSales` as if it were a regular table. It selects the `ProductID` and `TotalSalesAmount` from this temporary result set and filters for products where `TotalSalesAmount` is greater than 1000.

This example shows how the CTE makes the query more modular by first encapsulating the aggregation logic and then using that aggregate result in a subsequent filtering step.

# **Question 5 : Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).**

### **Concept of SQL Normalization**

SQL Normalization is a process used in relational database design to organize the columns and tables of a relational database to minimize data redundancy and improve data integrity. It involves breaking down large tables into smaller, more focused tables and defining relationships between them.

### **Primary Goals of Normalization:**

1.  **Eliminate Redundant Data**: Reduce the duplication of data stored in the database. This saves storage space and prevents inconsistencies that can arise when the same data is stored in multiple places.
2.  **Improve Data Integrity**: Ensure that data is accurate and consistent across the database. This means that changes to data only need to be made in one place, reducing the risk of errors.
3.  **Reduce Data Anomalies**: Prevent insertion, update, and deletion anomalies:
    *   **Insertion Anomaly**: Difficulty inserting new data into the database if other related data is not yet available.
    *   **Update Anomaly**: The need to update multiple rows to reflect a single change, potentially leading to inconsistencies if all copies are not updated.
    *   **Deletion Anomaly**: The loss of important data when a record is deleted because it was the only source of that information.
4.  **Simplify Database Design**: Create a more logical and organized database structure, making it easier to understand, maintain, and extend.

### **Brief Explanation of the First Three Normal Forms:**

Database normalization is typically carried out in stages, with each stage (or normal form) enforcing a stricter set of rules.

1.  **First Normal Form (1NF)**
    *   **Rule**: A table is in 1NF if it meets the following conditions:
        *   Each column contains atomic (indivisible) values. There should be no multi-valued attributes in a single cell.
        *   Each row is unique, meaning there's a primary key that uniquely identifies each record.
    *   **Goal**: To eliminate repeating groups within rows.
    *   **Example (Not 1NF)**:
        ```
        OrderID | CustomerName | ProductNames
        -------------------------------------
        1       | Alice        | Laptop, Mouse
        2       | Bob          | Keyboard
        ```
    *   **Example (1NF)**:
        ```
        OrderID | CustomerName | ProductName
        -----------------------------------
        1       | Alice        | Laptop
        1       | Alice        | Mouse
        2       | Bob          | Keyboard
        ```

2.  **Second Normal Form (2NF)**
    *   **Rule**: A table is in 2NF if it is in 1NF **and** all non-key attributes are fully functionally dependent on the primary key. This means that no non-key attribute is dependent on only a *part* of a composite primary key.
    *   **Goal**: To eliminate partial dependencies.
    *   **Scenario**: Consider a table with a composite primary key `(OrderID, ProductID)` and attributes `ProductName`, `OrderDate`.
        *   `ProductName` is dependent only on `ProductID` (part of the primary key).
        *   `OrderDate` is dependent only on `OrderID` (part of the primary key).
    *   **Example (Not 2NF)**:
        ```
        (OrderID, ProductID) | OrderDate  | ProductName
        -----------------------------------------------
        (1, A)               | 2023-01-01 | Laptop
        (1, B)               | 2023-01-01 | Mouse
        (2, C)               | 2023-01-02 | Keyboard
        ```
    *   **Example (2NF - decomposed)**:
        `Orders` table:
        ```
        OrderID | OrderDate
        -------------------
        1       | 2023-01-01
        2       | 2023-01-02
        ```
        `Order_Items` table:
        ```
        OrderID | ProductID | ProductName
        ----------------------------------
        1       | A         | Laptop
        1       | B         | Mouse
        2       | C         | Keyboard
        ```

3.  **Third Normal Form (3NF)**
    *   **Rule**: A table is in 3NF if it is in 2NF **and** all non-key attributes are non-transitively dependent on the primary key. This means there are no transitive dependencies, where a non-key attribute is dependent on another non-key attribute.
    *   **Goal**: To eliminate transitive dependencies.
    *   **Scenario**: Consider an `Employees` table with `EmployeeID` as primary key, `DepartmentID`, and `DepartmentName`.
        *   `EmployeeID` -> `DepartmentID`
        *   `DepartmentID` -> `DepartmentName` (a non-key attribute `DepartmentName` depends on another non-key attribute `DepartmentID`)
    *   **Example (Not 3NF)**:
        ```
        EmployeeID | EmployeeName | DepartmentID | DepartmentName
        ---------------------------------------------------------
        1          | John Doe     | 101          | Sales
        2          | Jane Smith   | 102          | Marketing
        ```
    *   **Example (3NF - decomposed)**:
        `Employees` table:
        ```
        EmployeeID | EmployeeName | DepartmentID
        ---------------------------------------
        1          | John Doe     | 101
        2          | Jane Smith   | 102
        ```
        `Departments` table:
        ```
        DepartmentID | DepartmentName
        ------------------------------
        101          | Sales
        102          | Marketing
        ```

````python
import sqlite3

# Connect to an in-memory SQLite database
# This creates a database that exists only for the duration of the session
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

print("Database 'ECommerceDB' created (in-memory).")
````

### Creating Tables

````python
# Create Categories table
cursor.execute('''
CREATE TABLE Categories (
    CategoryID INTEGER PRIMARY KEY,
    CategoryName VARCHAR(50) NOT NULL UNIQUE
);
''')
print("Table 'Categories' created.")

# Create Products table
cursor.execute('''
CREATE TABLE Products (
    ProductID INTEGER PRIMARY KEY,
    ProductName VARCHAR(100) NOT NULL UNIQUE,
    CategoryID INTEGER,
    Price DECIMAL(10,2) NOT NULL,
    StockQuantity INTEGER,
    FOREIGN KEY (CategoryID) REFERENCES Categories(CategoryID)
);
''')
print("Table 'Products' created.")

# Create Customers table
cursor.execute('''
CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY,
    CustomerName VARCHAR(100) NOT NULL,
    Email VARCHAR(100) UNIQUE,
    JoinDate DATE
);
''')
print("Table 'Customers' created.")

# Create Orders table
cursor.execute('''
CREATE TABLE Orders (
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    OrderDate DATE NOT NULL,
    TotalAmount DECIMAL(10,2),
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);
''')
print("Table 'Orders' created.")

conn.commit()
print("All tables created successfully.")
````

# **Question 7 : Generate a report showing CustomerName, Email, and the TotalNumberofOrders for each customer. Include customers who have not placed any orders, in which case their TotalNumberofOrders should be 0. Order the results by CustomerName**

````python
print("\n--- Report: Customer Orders ---")

cursor.execute('''
SELECT
    c.CustomerName,
    c.Email,
    COUNT(o.OrderID) AS TotalNumberOfOrders
FROM
    Customers AS c
LEFT JOIN
    Orders AS o ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID, c.CustomerName, c.Email
ORDER BY
    c.CustomerName;
''')

report_rows = cursor.fetchall()

# Print header
print(f"{"Customer Name":<20} {"Email":<25} {"Total Orders":<15}")
print("-" * 60)

# Print data
for row in report_rows:
    print(f"{row[0]:<20} {row[1]:<25} {row[2]:<15}")
````

# **Question 8 : Retrieve Product Information with Category: Write a SQL query to display the ProductName, Price, StockQuantity, and CategoryName for all products. Order the results by CategoryName and then ProductName alphabetically.**

````python
print("\n--- Report: Product Information with Categories ---")

cursor.execute('''
SELECT
    p.ProductName,
    p.Price,
    p.StockQuantity,
    c.CategoryName
FROM
    Products AS p
JOIN
    Categories AS c ON p.CategoryID = c.CategoryID
ORDER BY
    c.CategoryName, p.ProductName;
''')

product_report_rows = cursor.fetchall()

# Print header
print(f"{"Product Name":<30} {"Price":<10} {"Stock":<10} {"Category Name":<20}")
print("-" * 72)

# Print data
for row in product_report_rows:
    print(f"{row[0]:<30} {row[1]:<10.2f} {row[2]:<10} {row[3]:<20}")
````

# **Question 9 : Write a SQL query that uses a Common Table Expression (CTE) and a Window Function (specifically ROW_NUMBER() or RANK()) to display the CategoryName, ProductName, and Price for the top 2 most expensive products in each CategoryName.**

````python
print("\n--- Report: Top 2 Most Expensive Products per Category ---")

cursor.execute('''
WITH RankedProducts AS (
    SELECT
        p.ProductID,
        p.ProductName,
        p.Price,
        p.CategoryID,
        c.CategoryName,
        ROW_NUMBER() OVER (PARTITION BY p.CategoryID ORDER BY p.Price DESC) as rn
    FROM
        Products AS p
    JOIN
        Categories AS c ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM
    RankedProducts
WHERE
    rn <= 2
ORDER BY
    CategoryName, Price DESC;
''')

report_rows_q9 = cursor.fetchall()

# Print header
print(f"{"Category Name":<20} {"Product Name":<30} {"Price":<10}")
print("-" * 60)

# Print data
for row in report_rows_q9:
    print(f"{row[0]:<20} {row[1]:<30} {row[2]:<10.2f}")
````

# **Question 10 : You are hired as a data analyst by Sakila Video Rentals, a global movie rental company. The management team is looking to improve decision-making by analyzing existing customer, rental, and inventory data.**

**Using the Sakila database, answer the following business questions to support key strategic initiatives.**

**Tasks & Questions:**
1.  Identify the top 5 customers based on the total amount they’ve spent. Include customer name, email, and total amount spent.
2.  Which 3 movie categories have the highest rental counts? Display the category name and number of times movies from that category were rented.
3.  Calculate how many films are available at each store and how many of those have never been rented.
4.  Show the total revenue per month for the year 2023 to analyze business seasonality.
5.  Identify customers who have rented more than 10 times in the last 6 months.

### Setting up a dummy Sakila-like SQLite Database

````python
import sqlite3

# Connect to an in-memory SQLite database
sakila_conn = sqlite3.connect(':memory:')
sakila_cursor = sakila_conn.cursor()

print("In-memory Sakila-like database created.")
````

### Creating Sakila-like Tables

````python
# Create Customer table
sakila_cursor.execute('''
CREATE TABLE Customer (
    customer_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    email TEXT UNIQUE
);
''')

# Create Film table
sakila_cursor.execute('''
CREATE TABLE Film (
    film_id INTEGER PRIMARY KEY,
    title TEXT NOT NULL
);
''')

# Create Category table
sakila_cursor.execute('''
CREATE TABLE Category (
    category_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL UNIQUE
);
''')

# Create Film_Category table (many-to-many relationship)
sakila_cursor.execute('''
CREATE TABLE Film_Category (
    film_id INTEGER,
    category_id INTEGER,
    PRIMARY KEY (film_id, category_id),
    FOREIGN KEY (film_id) REFERENCES Film(film_id),
    FOREIGN KEY (category_id) REFERENCES Category(category_id)
);
''')

# Create Store table
sakila_cursor.execute('''
CREATE TABLE Store (
    store_id INTEGER PRIMARY KEY,
    address TEXT NOT NULL
);
''')

# Create Inventory table
sakila_cursor.execute('''
CREATE TABLE Inventory (
    inventory_id INTEGER PRIMARY KEY,
    film_id INTEGER NOT NULL,
    store_id INTEGER NOT NULL,
    FOREIGN KEY (film_id) REFERENCES Film(film_id),
    FOREIGN KEY (store_id) REFERENCES Store(store_id)
);
''')

# Create Rental table
sakila_cursor.execute('''
CREATE TABLE Rental (
    rental_id INTEGER PRIMARY KEY,
    rental_date DATETIME NOT NULL,
    inventory_id INTEGER NOT NULL,
    customer_id INTEGER NOT NULL,
    FOREIGN KEY (inventory_id) REFERENCES Inventory(inventory_id),
    FOREIGN KEY (customer_id) REFERENCES Customer(customer_id)
);
''')

# Create Payment table
sakila_cursor.execute('''
CREATE TABLE Payment (
    payment_id INTEGER PRIMARY KEY,
    customer_id INTEGER NOT NULL,
    rental_id INTEGER,
    amount DECIMAL(5,2) NOT NULL,
    payment_date DATETIME NOT NULL,
    FOREIGN KEY (customer_id) REFERENCES Customer(customer_id),
    FOREIGN KEY (rental_id) REFERENCES Rental(rental_id)
);
''')

sakila_conn.commit()
print("All Sakila-like tables created successfully.")
````

### Answering Question 10.5: Customers with More Than 10 Rentals in Last 6 Months

````python
print("\n--- Report: Customers with >10 Rentals in Last 6 Months ---")

sakila_cursor.execute('''
SELECT
    c.first_name || ' ' || c.last_name AS customer_name,
    c.email,
    COUNT(r.rental_id) AS rental_count
FROM
    Customer AS c
JOIN
    Rental AS r ON c.customer_id = r.customer_id
WHERE
    r.rental_date >= DATE('now', '-6 months')
GROUP BY
    c.customer_id, customer_name, c.email
HAVING
    COUNT(r.rental_id) > 10
ORDER BY
    rental_count DESC;
''')

loyal_customers_report = sakila_cursor.fetchall()

# Print header
print(f"{"Customer Name":<25} {"Email":<30} {"Rental Count":<15}")
print("-" * 70)

# Print data
if not loyal_customers_report:
    print("No customers found with more than 10 rentals in the last 6 months.")
else:
    for row in loyal_customers_report:
        print(f"{row[0]:<25} {row[1]:<30} {row[2]:<15}")
````

### Answering Question 10.4: Total Revenue per Month for 2023

````python
print("\n--- Report: Total Revenue per Month for 2023 ---")

sakila_cursor.execute('''
SELECT
    STRFTIME('%Y-%m', payment_date) AS sales_month,
    SUM(amount) AS total_revenue
FROM
    Payment
WHERE
    STRFTIME('%Y', payment_date) = '2023'
GROUP BY
    sales_month
ORDER BY
    sales_month;
''')

monthly_revenue_report = sakila_cursor.fetchall()

# Print header
print(f"{"Month":<10} {"Total Revenue":<15}")
print("-" * 25)

# Print data
for row in monthly_revenue_report:
    print(f"{row[0]:<10} {row[1]:<15.2f}")
````

### Answering Question 10.3: Films Available at Each Store and Never Rented

````python
print("\n--- Report: Films Availability and Never Rented per Store ---")

sakila_cursor.execute('''
SELECT
    s.address AS store_address,
    COUNT(i.inventory_id) AS total_films_available,
    SUM(CASE WHEN r.rental_id IS NULL THEN 1 ELSE 0 END) AS films_never_rented
FROM
    Store AS s
JOIN
    Inventory AS i ON s.store_id = i.store_id
LEFT JOIN
    Rental AS r ON i.inventory_id = r.inventory_id
GROUP BY
    s.store_id, s.address
ORDER BY
    s.store_id;
''')

film_availability_report = sakila_cursor.fetchall()

# Print header
print(f"{"Store Address":<30} {"Total Films Available":<25} {"Films Never Rented":<20}")
print("-" * 75)

# Print data
for row in film_availability_report:
    print(f"{row[0]:<30} {row[1]:<25} {row[2]:<20}")
````

### Answering Question 10.2: Top 3 Movie Categories by Rental Counts

````python
print("\n--- Report: Top 3 Movie Categories by Rental Counts ---")

sakila_cursor.execute('''
SELECT
    cat.name AS category_name,
    COUNT(r.rental_id) AS rental_count
FROM
    Category AS cat
JOIN
    Film_Category AS fc ON cat.category_id = fc.category_id
JOIN
    Film AS f ON fc.film_id = f.film_id
JOIN
    Inventory AS i ON f.film_id = i.film_id
JOIN
    Rental AS r ON i.inventory_id = r.inventory_id
GROUP BY
    cat.name
ORDER BY
    rental_count DESC
LIMIT 3;
''')

top_categories = sakila_cursor.fetchall()

# Print header
print(f"{"Category Name":<25} {"Rental Count":<15}")
print("-" * 40)

# Print data
for row in top_categories:
    print(f"{row[0]:<25} {row[1]:<15}")
````

### Answering Question 10.1: Top 5 Customers by Total Amount Spent

````python
print("\n--- Report: Top 5 Customers by Total Amount Spent ---")

sakila_cursor.execute('''
SELECT
    c.first_name || ' ' || c.last_name AS customer_name,
    c.email,
    SUM(p.amount) AS total_amount_spent
FROM
    Customer AS c
JOIN
    Payment AS p ON c.customer_id = p.customer_id
GROUP BY
    c.customer_id, customer_name, c.email
ORDER BY
    total_amount_spent DESC
LIMIT 5;
''')

top_customers = sakila_cursor.fetchall()

# Print header
print(f"{"Customer Name":<25} {"Email":<30} {"Total Amount Spent":<20}")
print("-" * 75)

# Print data
for row in top_customers:
    print(f"{row[0]:<25} {row[1]:<30} {row[2]:<20.2f}")
````
